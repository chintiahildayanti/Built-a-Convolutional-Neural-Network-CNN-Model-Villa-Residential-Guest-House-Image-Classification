## 2. Convert Images (webp to jpg)

```python
import os
from PIL import Image

# Folder utama tempat file berada
folder_path = "/content/drive/My Drive/bukit_vista_images"  # Ganti dengan lokasi folder utama

# Loop untuk membaca semua file dalam semua subfolder
for root, _, files in os.walk(folder_path):
    for filename in files:
        if filename.lower().endswith(".webp"):
            webp_path = os.path.join(root, filename)
            jpg_path = os.path.join(root, filename.replace(".webp", ".jpg"))

            # Buka dan konversi gambar
            img = Image.open(webp_path)
            img.convert("RGB").save(jpg_path, "JPEG")

            # Hapus file .webp setelah konversi sukses
            os.remove(webp_path)

            print(f"Konversi {filename} selesai di {root}!")

print("Semua file .webp dalam folder dan subfolder telah dikonversi ke .jpg!")
```
